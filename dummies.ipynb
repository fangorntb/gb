```python
import random
import pandas as pd

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})
one_hot = pd.get_dummies(data, columns=['whoAmI'])
one_hot.head()
```

```python
labels = data['whoAmI'].unique()
one_hot_dict = {f'whoAmI_{label}': (data['whoAmI'] == label).astype(int) for label in labels}
one_hot_df = pd.DataFrame(one_hot_dict)
data = pd.concat([data, one_hot_df], axis=1)
data = data.drop('whoAmI', axis=1)
data.head()
```
